# Flight Fare prediction

# required liabraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import datetime
import time
import warnings
warnings.filterwarnings('ignore')
```

# Problem statement

```python
# From given data predict the price of Flight using random forest Regresion
```

# Data Gathering

```python
path=r"Data_Train.xlsx"
df_train=pd.read_excel(path)
df_train
```

```python
df_train.info()
```

```python
df_train
```

```python
airline=pd.get_dummies(df_train['Airline'],columns=True,drop_first=True)
```

```python
df_train=pd.concat([df_train,airline],axis=1)
```

```python
df_train["Total_Stops"].replace({'1 stop':1,
 'non-stop': 0,
 '2 stops': 2,
 '3 stops': 3,
 '4 stops': 4},inplace=True)
```

```python
df_train.drop(["Route","Airline"],axis=1,inplace=True)
```

```python
df_train.drop("Additional_Info",axis=1,inplace=True)
```

```python
df_train["Source"].value_counts().to_dict()
```

```python
df_train["Source"].replace({'Delhi': 0,
 'Kolkata': 1,
 'Banglore': 2,
 'Mumbai': 3,
 'Chennai': 4},inplace=True)
```

```python
df_train["Destination"].value_counts().to_dict()
```

```python
df_train["Destination"].replace({'Cochin': 5,
 'Banglore': 2,
 'Delhi': 0,
 'New Delhi': 6,
 'Hyderabad': 7,
 'Kolkata': 1},inplace=True)
```

```python
df_train["Date_of_Journey"]=df_train["Date_of_Journey"].astype(str)
```

```python
df_train["Date_of_Journey"].dtype
```

```python
pattern="%d/%m/%Y"
Day=[]
Month=[]
Year=[]
for i in df_train["Date_of_Journey"]:
    date=datetime.datetime.strptime(i,pattern)
    Day.append(date.date().day)
    Month.append(date.date().month)
Day
Month
Year
```

```python
df_train["Journey_Month"]=Month
df_train["Journey_Day"]=Day
```

```python
df_train.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df_train.drop("Arrival_Time",axis=1,inplace=True)
```

```python
Hours=[]
Minute=[]
patt="%H:%M"
for i in df_train["Dep_Time"]:
    date=datetime.datetime.strptime(i,patt)
    Hours.append(date.hour)
    Minute.append(date.minute)
Hours
Minute
```

```python
df_train["Dep_Time_Hours"]=Hours
df_train["Dep_Time_Minute"]=Minute
```

```python
df_train.fillna(df_train["Total_Stops"].mode()[0],inplace=True)
```

```python
duration=list(df_train["Duration"])
for i in range(len(df_train["Duration"])):
    if len(duration[i].split())!=2:
        if "h" in duration[i]:
            duration[i]=duration[i].strip()+" 0m"
        else:
            duration[i]="0h "+duration[i]
duration_h=[]
duration_m=[]
for i in range (len(duration)):
    duration_h.append(int(duration[i].split(sep="h")[0]))
    duration_m.append(int(duration[i].split(sep="m")[0].split()[-1]))
print(duration_h)
print(duration_m)
```

```python
df_train["Duration_hours"]=duration_h
df_train["Duration_min"]=duration_m
```

```python
df_train.drop(["Dep_Time","Duration"],axis=1,inplace=True)
```

```python
df_train[["Total_Stops"]]=df_train[["Total_Stops"]].median()
```

```python
df_train[["Duration_hours"]].mean()
```

```python
df_train
```

```python
df_train.info()
```

```python
df_train.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_train.corr(),annot=True)
```

```python
x=df_train.drop("Price",axis=1)
y=df_train["Price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=5)
```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
```

```python
hyp={'n_estimators':np.arange(10,200,5),
    'criterion':['squared_error','absolute_error'],
    'max_depth':np.arange(1,20),
    'min_samples_split':np.arange(1,20),
    'min_samples_leaf':np.arange(1,20)}
gscv=RandomizedSearchCV(rfr,hyp,cv=5)
gscv.fit(x_train,y_train)
```

```python
gscv.best_params_
```

```python
rfr=RandomForestRegressor(n_estimators=85,min_samples_split=25,min_samples_leaf=1,max_depth=22,criterion='squared_error')
rfr.fit(x_train,y_train)
```

```python
pred_train=rfr.predict(x_train)
pred_train
```

```python
mse=mean_squared_error(y_train,pred_train)
print(mse)
rmse=np.sqrt(mse)
print(rmse)
mae=mean_absolute_error(y_train,pred_train)
print(mae)
r2=r2_score(y_train,pred_train)
print(r2)
```

```python
pred_test=rfr.predict(x_test)
pred_test
```

```python
mse=mean_squared_error(y_test,pred_test)
print(mse)
rmse=np.sqrt(mse)
print(rmse)
mae=mean_absolute_error(y_test,pred_test)
print(mae)
r2=r2_score(y_test,pred_test)
print(r2)
```
